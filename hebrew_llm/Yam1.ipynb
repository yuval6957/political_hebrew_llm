```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()
```

```python
from transformers import pipeline
import torch
from transformers import logging
from tqdm import tqdm
logging.set_verbosity_error()
```

```python
gc=gspread.authorize(creds)
complition_sh=gc.open('complition_questions').sheet1
evaluation_sh=gc.open('Hebrew_Mistral_7B').sheet1
```

```python
generator=pipeline('text-generation', 'yam-peleg/Hebrew-Mistral-7B')
```

```python
sentences = [x[0] for x in complition_sh.get('A2:100')]
print('\n'.join(sentences))
for row,sentence in enumerate(tqdm(sentences)):
  if len(evaluation_sh.get(f'A{row+2}')[0])>0:
    continue
  prompt = sentence
  num_sequences = 5
  result = generator(prompt.strip(), do_sample=True, max_new_tokens=100, return_full_text=False,temperature=0.5,num_return_sequences=num_sequences)
  for i in range(num_sequences):
    generated_text = result[i]['generated_text']
    evaluation_sh.update_cell(row+2, i+1, generated_text)
    print(prompt+generated_text )
    print('----------------')
```
